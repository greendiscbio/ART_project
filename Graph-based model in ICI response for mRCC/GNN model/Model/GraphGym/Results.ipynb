```python
import json
import shutil
import numpy as np
import os
i = 1
final_val_f1 = []
final_train_f1 = []
final_train_accuracy = []
final_val_accuracy = []
final_train_precision = []
final_val_precision = []
final_train_recall = []
final_val_recall =[]

for dir in listdirs(path):
  print(dir)
  if os.path.exists(path + '/'+str(dir)+'/agg'):
    shutil.rmtree(path + '/'+str(dir)+'/agg')
  for subdir in listdirs(path+'/'+str(dir)):
    train_f1 = []
    val_f1 =[]
    train_accuracy = []
    val_accuracy =[]
    train_precision = []
    val_precision =[]
    train_recall = []
    val_recall =[]
    for subsubdir in listdirs(path+'/'+str(dir)+'/'+str(subdir)): 
      if os.path.exists(path+'/'+str(dir)+'/'+str(subdir)+'/'+str(subsubdir)+'/train/stats.json'):   
        with open(path+'/'+str(dir)+'/'+str(subdir)+'/'+str(subsubdir)+'/train/stats.json') as file:
          last_line = file.readlines()[-1]
          data = json.loads(last_line)
          train_f1.append(data["f1"])
          train_accuracy.append(data["accuracy"])
          train_precision.append(data["precision"])
          train_recall.append(data["recall"])
        with open(path+'/'+str(dir)+'/'+str(subdir)+'/'+str(subsubdir)+'/val/stats.json') as file:
          last_line = file.readlines()[-1]
          data = json.loads(last_line)
          val_f1.append(data["f1"])
          val_accuracy.append(data["accuracy"])
          val_precision.append(data["precision"])
          val_recall.append(data["recall"])
      else:
        print('Not found'+ path+'/'+str(dir)+'/'+str(subdir)+'/'+str(subsubdir)+'/train/stats.json')
    if os.path.exists(path+'/'+str(dir)+'/'+str(subdir)+'/'+str(subsubdir)+'/train/stats.json'):
      final_val_f1.append(np.array(val_f1).mean())
      final_val_precision.append(np.array(val_precision).mean())
      final_val_recall.append(np.array(val_recall).mean())
      final_val_accuracy.append(np.array(val_accuracy).mean())
      final_train_f1.append(np.array(train_f1).mean())
      final_train_precision.append(np.array(train_precision).mean())
      final_train_recall.append(np.array(train_recall).mean())
      final_train_accuracy.append(np.array(train_accuracy).mean())

      
  print('------------------- Split '+str(i)+' -------------------')
  print('Number of arquitectures: '+str(len(final_val_f1)))
    # print('Mean F1-score in train: ' + str(np.numpy(final_val).mean()))
    # print('Number of arquitectures: '+str(len(final_val)))
    # print('Mean F1-score in test: ' + str(np.numpy(final_val).mean()))
  i = i+1
```
