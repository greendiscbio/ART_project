```python
df = pd.read_csv('clinic_and_RNA_data_raw_Ave_Axi.csv')
df.drop(columns = ['Unnamed: 0'], inplace=True)
df.rename(columns = {"PFS_P": "PFS"}, inplace = True)
df
```

# Sunitinib

## Read data

```python
df = pd.read_csv('clinic_and_RNA_data_raw_Sunitinib.csv')
df.drop(columns = ['Unnamed: 0'], inplace=True)
df.rename(columns = {"PFS_P": "PFS"}, inplace = True)
df
```

## Missing values analysis

```python
((df.isnull().sum() / len(df))*100).sort_values(ascending = False)
```

```python
df.dropna(axis=1, thresh=len(df),inplace=True)
df
```

```python
((df.isnull().sum() / len(df))*100).sort_values(ascending = False)
```

## Data type

```python
df.dtypes
```

## Data set statistics

```python
print('Mean: '+str(float(df['PFS'].mean())))
print('Median: '+str(float(df['PFS'].median())))
```
