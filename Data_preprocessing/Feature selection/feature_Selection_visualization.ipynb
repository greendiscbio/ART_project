```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path ='150_feature_selection.csv'
data = pd.read_csv(path)
data.head(5)
```

```python
import re
```

```python
cvs = []
for i in data.cv_scores:
    patn = re.sub(r"[\([{})\]]", "", i)
    patn=round(float(patn),2)
    cvs.append(patn)
```

```python
cvs
```

```python
plt.plot(cvs, label='Biderectional elimination')
plt.plot(96.5,1, marker="x", markersize=7,color="red" ,label='sticking point (98 features)')
plt.plot(28.5,0.76, marker="x",  markersize=7 , color="purple" ,label='sticking point (30 features)')
plt.plot(13.5,0.5, marker="x",  markersize=7, color="green" ,label='sticking point (15 features)')

plt.legend()
plt.show()
```
