# Nivolumab

## Read data

```python
df = pd.read_csv('clinic_and_RNA_data_raw_NIVOLUMAB.csv')
df.drop(columns = ['Unnamed: 0'], inplace=True)
df
```

```python
df.describe()
```

## Missing values analysis

```python
((df.isnull().sum() / len(df))*100).sort_values(ascending = False)
```

```python
df.dropna(axis=1, thresh=(0.8*len(df)),inplace=True)
df
```

```python
((df.isnull().sum() / len(df))*100).sort_values(ascending = False)
```

## Data type

```python
df.dtypes
```

## Dataset statistics

```python
print('Mean: '+str(float(df['PFS'].mean())))
print('Median: '+str(float(df['PFS'].median())))
print('Patients with PFS < '+str(float(df['PFS'].median())) + ' months: '+str(len(df[ df['PFS']<float(df['PFS'].median())])/len(df))+'%')
```

```python
print('Patients with PFS < 3 months: '+str(len(df[ df['PFS']<3.5])/len(df))+'%')
# print('Patients with PFS between 3 and 6 months: '+str(len(df[ df['PFS'].between(3,6)])/len(df))+'%')
print('Patients with PFS > 6 months: '+str(len(df[ df['PFS']>3.5])/len(df))+'%')
```
