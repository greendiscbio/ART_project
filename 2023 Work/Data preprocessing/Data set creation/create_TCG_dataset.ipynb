```python
data_clinical_patient =pd.read_csv('2023 Work/Data/kirc_tcga_pan_can_atlas_2018/data_clinical_patient.txt', sep='\t')
data_clinical_sample =pd.read_csv('2023 Work/Data/kirc_tcga_pan_can_atlas_2018/data_clinical_sample.txt', sep='\t')
```

```python
data_clinical_sample
```

```python
data_clinical_patient
```

```python
clinical_data = data_clinical_patient.merge(data_clinical_sample, on='#Patient Identifier')
```

```python
clinical_data
```

```python
clinical_data.drop([0,1,2,3], inplace=True)
clinical_data
```

Identify and calculate the percentage of the missing values in each attribute

```python
clinical_data.isnull().sum()/clinical_data.shape[0]*100
```

Drop columns which have more than 60% of NaN

```python
clinical_data = clinical_data.dropna(thresh=clinical_data.shape[0]*0.6,how='all',axis=1)
clinical_data
```

Drop columns which have the same values in all rows

```python
nunique = clinical_data.nunique()
cols_to_drop = nunique[nunique == 1].index
clinical_data.drop(cols_to_drop, axis=1, inplace=True)
clinical_data
```

```python
clinical_data.dtypes
```

```python
clinical_data.to_excel('clinical_data.xlsx')
```

```python
print('Patients whose PFS is over 6 months: '+str(len(clinical_data[clinical_data['Progress Free Survival (Months)'].astype('float')>6])))
print('Patients whose PFS is between 3 and 6 months: '+str(len(clinical_data[clinical_data['Progress Free Survival (Months)'].astype('float').between(3,6)])))
print('Patients whose PFS is lower 3 months: '+str(len(clinical_data[clinical_data['Progress Free Survival (Months)'].astype('float')<3])))
```
